```python
from  selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from collections import defaultdict # Default dictionary: store a list with each key
import re
```

```python
options = Options()
options.add_argument("--disable-extensions")
driver = webdriver.Chrome(options=options)
```

```python
#indeed_posts=[]
# create dictionary
indeed_posts={"JobTitle":[],
     "CompanyName":[],
     "Salary":[],
     "Location":[],
     "Link":[],
     "Salary":[],
     "PostJob":[],
     "PostJobDay":[]
    
}
```

```python
""""no_of_pages=2
skill='Electrician'
#places=["Noida","Faridabad","Delhi","Gurgaon","Sahibabad","Gurugram","Haryana","Greater Noida","Ghaziabad"]
#places=["Noida","Faridabad"]
for place in places:
  
"""
places=["Noida","Faridabad","Delhi","Gurgaon","Sahibabad","Gurugram","Haryana","Greater Noida","Ghaziabad"]
no_of_pages=10
skill="Carpenter"
for place in places:
    for page in range(no_of_pages):
        options.add_argument("--disable-extensions")
        url = 'https://www.indeed.co.in/jobs?q=' + skill + \
            '&l=' + place + '&sort=date' +'&start='+ str(page * 10)
        driver.get(url)
        all_jobs=driver.find_elements_by_class_name('result')
        #print(all_jobs)
        for job in all_jobs:
            result_html=job.get_attribute('innerHTML')
            soup=BeautifulSoup(result_html,'html.parser')
            # print(soup)
    
            outer_most_point=soup.find('div',attrs={'class': 'job_seen_beacon'})
            #print(outer_most_point)
            
            for i in outer_most_point.find('table',attrs={'class': 'jobCard_mainContent big6_visualChanges'}):
                #print(i)
                job_title=i.find('h2',{'class':'jobTitle css-1h4a4n5 eu4oa1w0'})
                #print(job_title)
                if job_title!=None:
                    jobs=job_title.find('a').text
                    #print(jobs)
                #Company Name:
                if i.find('span',{'class':'companyName'})!=None:
                    company=i.find('span',{'class':'companyName'}).text
                    #print(company)
               
                #Location:
                if i.find('div',{'class':'companyLocation'})!=None:
                    location=i.find('div',{'class':'companyLocation'}).text
                    
                # Links: these Href links will take us to full job description
                if i.find('a') != None:
                    links=i.find('a',{'class':'jcs-JobTitle css-jspxzf eu4oa1w0'})['href']
                    links='https://in.indeed.com'+links
                    #print(links)
                
                # Salary if available:
                if i.find('div',{'class':'attribute_snippet'}) != None:
                    salary=i.find('div',{'class':'attribute_snippet'}).text

                else:
                    salary='No Salary'
            
                for i in outer_most_point.find('table',attrs={'class': 'jobCardShelfContainer big6_visualChanges'}):
                    # Job Post Date:
                    if i.find('span', attrs={'class': 'date'}) != None:
                        post_date = i.find('span', attrs={'class': 'date'}).text
                        #print(str(post_date))
                        temp = re.findall(r'\d+',str(post_date))
                        for i in temp:
                            postdate=i
                        #print(type(temp))
                indeed_posts['JobTitle'].append(jobs)
                indeed_posts['CompanyName'].append(company)
                indeed_posts['Location'].append(location)
                indeed_posts['Link'].append(links)
                indeed_posts['Salary'].append(salary)
                indeed_posts['PostJob'].append(post_date)
                indeed_posts['PostJobDay'].append(postdate)
```
